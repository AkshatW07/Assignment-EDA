```python
summary = pd.DataFrame({'Mean':df.mean(),'Median':df.median(),'Std_dev':df.std(),'IQR':df.quantile(0.75)-df.quantile(0.25)})
summary
```

```python
### distribution of variables
df.hist(figsize=(15,10),bins=30)
plt.tight_layout()
plt.show()
```

```python
### boxplots for spreads and outliers
plt.figure(figsize=(15,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
### correlation heatmap
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
### ASTV negatively correlates with MSTV
### ALTV and MLTV show relationship
### Strong correlations indicate redundant features
```

```python
### pairplot
sns.pairplot(df[['LB','ASTV','MSTV','ALTV','MLTV']])
plt.show()
```

```python
### Useful for spotting non-linear relationships and clusters.
```

```python
### violin plot
plt.figure(figsize=(10,6))
sns.violinplot(data=df[['ASTV','MSTV','ALTV','MLTV']])
plt.show()
```

```python
### Key Patterns Identified:

### 1 High ASTV & ALTV  Abnormal fetal heart rate variability

### 2 Low MSTV & MLTV  Reduced variability  possible distress

### 3 Decelerations (DL, DP) occur infrequently but are clinically significant

### 4 Many variables are highly skewed, suggesting need for scaling or transformation before modeling
```

 Recommendations for Further Analysis

Apply log transformation on skewed variables

Use PCA or correlation-based feature selection

Build classification models (Normal vs Abnormal FHR)
