```python
### scalling techniques
###numerical featrures
num_cols = df.select_dtypes(include=['int64','float64']).columns
```

```python
### standard scalling
from sklearn.preprocessing import StandardScaler

scaler_std= StandardScaler()
df_std_scaled = df.copy()
df_std_scaled[num_cols] = scaler_std.fit_transform(df[num_cols])
```

```python
###min-max scalling
from sklearn.preprocessing import MinMaxScaler

scaler_mm = MinMaxScaler()
df_mm_scaled = df.copy()
df_mm_scaled[num_cols]=scaler_mm.fit_transform(df[num_cols])
```

### WHEN TO USE WHICH

Standard Scaling - ML algorithms assuming normal distribution (Logistic Regression, SVM, Linear Regression)
Min-Max Scaling - Distance-based models (KNN, Neural Networks) & bounded inputs

```python
###encoding techniques
#seprate cat columns
cat_cols = df.select_dtypes(include='object').columns
```

```python
###One-Hot Encoding (< 5 categories)
low_cardinality=[col for col in cat_cols if df[col].nunique() < 5]

df_ohe = pd.get_dummies(df,columns=low_cardinality,drop_first=True)
```

```python
### Label Encoding (> 5 categories)

from sklearn.preprocessing import LabelEncoder

high_cardinality = [col for col in cat_cols if col not in low_cardinality]

le = LabelEncoder()
for col in high_cardinality:
    df_ohe[col] = le.fit_transform(df_ohe[col])
```

```python
### feature engineering
# new feature 1:capital not gain
df_ohe['capital_net'] = df_ohe['capital_gain'] - df_ohe['capital_loss']
```

```python
### New Feature 2: Work Intensity
df_ohe['work_intensity'] = df_ohe['hours_per_week'] * df_ohe['education_num']
```

```python
### Feature Transformation (Skewness Handling)
df_ohe['capital_gain_log'] = np.log1p(df_ohe['capital_gain'])
```
