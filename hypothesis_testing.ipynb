```python
###1
print('H0: Mean Weekly operating cost = theoratical cost')
print('H1: Mean Weekly operating cost > theoratical cost')
```

```python
##2 calculate the test stats (z)
import numpy as np

fixed_cost = 1000
cost_per_unit = 5
mean_units = 600

mu = fixed_cost + cost_per_unit * mean_units
print('Theoratical mean weekly Cost',mu)
```

```python
sigma = 5 * 25
print('Standard Deviation of Weekly Cost:',sigma)
```

```python
### apply ztest formula
import math

sample_mean = 3050
n = 25

standard_error = sigma / math.sqrt(n)
z_stat = (sample_mean - mu)/standard_error

print('standard error:',standard_error)
print('Z statistic:',z_stat)
```

```python
###3
from scipy.stats import norm
alpha = 0.05
z_critical = norm.ppf(1-alpha)
p_value= 1 - norm.cdf(z_stat)

print('Critical Z value',z_critical)
print('p-value:',p_value)
```

```python
### 4 
if z_stat>z_critical:
    print('Reject The Null Hypothesis')
else:
    print('Fail to Reje')
```
